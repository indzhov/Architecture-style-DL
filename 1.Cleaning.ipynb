# A few words from the author: 
The objective of this task is pretty hard to be rached. The problems are numerous and we are going to try to solve them as much as possible. However, it is good to bear in mind that this is not a simple classification problem. For instance, we are not going to predict only if there is a building in a particular photo. In case to recognize the different architectural styles, our model have to look not only at the types, shapes, colors of the building, but also to keep attention to the different facades and ornaments. Not to mention, that some examples like the Egyptian ones, could not be classified as building at all, they have pyramids and statues of sphinxes. Even if we find a solution to recognize and classify those things, we wont have a valid classification in any case. Hypothetically, even when we can classify the visual appearance of the building on a hundred percent, it wont be true in the real life. This is because, apart from the style almost any civlization or country have contributed on every building with their own touches. Thus, a specific architecture could looks slightly different in two separated countries. Also, the historical time in which they were built also could impact slightly the visual appearance. Another problem are the influences. For example, we have "International style", which presents the buildings in the modern and big cities, like New York. Those buildings are usually skyscrapers. However,not only each of them is different to any other, but their heigh is varying in every country, city or district. Another examples, is the "revival" architectures. Those are buildings which were build more recently than the time from which the original style comes. Those kind of architectures could be influenced from the ancient civilizations and built nowadays. The Greeks influenced the Romans, which in turn brough their forms into the British Islands from where we can see their contribution all over the world. Thus, the Greek Revival's features could be seen not only in different places, from different historical time, but also in other styles that are not generally "Greek ones". There is also a problem with the refurbishment, or the damage done to some buildings through the ages. For instance, Hagia Sophia is perfect example of Byzantine architecture, but there is no even a single photo without the minarets, which are more common, for the Arabic architectures, but were added to Haghia Sophia, after the Ottomans seiged and took over the city of Constantinopol in 15th century. Then the building were changed from a Church to a Mosque. With all of that being said, it is clear that the task of classifying architectures equally and accuratelly become very hard. For that reason, a person with architectural, cultural or art historical knowledge is needed, especially in the preparing of the data from which our model will learn.

The papers itself is inspired from another findings. like the possibility  of  cross-domainmatching from sketches to building images. Zhe Xu used that, to solve the problem with representing features, like the building fa Ì§cades. The proposed algorithm, MLLR, uses latent variables. They could be use in other problems, like image captioning, where they are explained in the following way : "targeted partial caption in-dicates that the words at some locations are observed, whilethe words at other locations are not specified,i.e., latent" [1] The Latent variables arise mainly, but not exclusively, in the social sciences. This is because social science often deals in concepts which are constructs rather than the directly measurable variables which are typical of physical sciences. Clearly, they are variables that are not directly observed but are rather inferred (through a mathematical model) from other variables that are observed (directly measured). Thus, they can change their attitude (values) in different situations.
